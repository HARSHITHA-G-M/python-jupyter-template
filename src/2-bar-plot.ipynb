### Exercise 2: Bar Plot with Colors

```python
# TASK: Create a bar plot with the following data: categories = ['A', 'B', 'C', 'D'] and values = [5, 7, 3, 9].
# Use different colors for each bar and add a title to the plot.
# Import necessary libraries
import matplotlib.pyplot as plt

# Step 1: Define the categories and values
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]

# Step 2: Create a bar plot with different colors for each bar
plt.bar(categories, values, color=['red', 'green', 'blue', 'purple'])

# Step 3: Add a title to the plot
plt.title('Bar Plot of Categories vs Values')

# Step 4: Display the plot
plt.show()
```
